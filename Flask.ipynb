```python
# 1- Retrieve properties that have a price greater than 1 million and are located in "Estados Unidos" (l1).

query='''
select pp.id,pp.price,pd.l1
from property_price_details_table  as pp
join  property_details_table as pd
on pp.id=pd.id
where  pp.price > 1000000 and pd.l1='Estados Unidos';'''

results1=pd.read_sql(query,conn)
results1
```

```python
# 2- SQL query to categorize properties based on surface area
query = '''
SELECT *,
    CASE
        WHEN surface_total < 50 THEN 'Small'
        WHEN surface_total BETWEEN 50 AND 100 THEN 'Medium'
        WHEN surface_total > 100 THEN 'Large'
    END AS surface_total_category
FROM property_details_table;
'''

# Execute the query and fetch the results into a pandas DataFrame
result2 = pd.read_sql(query, conn)
result2
```

```python
# 3. List all properties (id) in the "Belgrano" neighborhood (l3) that have the same number of bedrooms and bathrooms as another property in the dataset:
query='''SELECT distinct p1.id,p1.bedrooms,p1.bathrooms
FROM property_details_table as p1
inner join
property_details_table as p2
on p1.bedrooms=p2.bathrooms
where p1.l3='Belgrano'
;'''

result3=pd.read_sql(query,conn)
result3
```

```python
# 4. Calculate the average price per square meter (price / surface_total) for each property type (property_type) in the "Belgrano" neighborhood (l3):

query='''SELECT avg(price/surface_total) as price_per_sq_meter , property_type
FROM property_details_table as pd
JOIN
property_price_details_table as pp
ON pd.id=pp.id
where l3='Belgrano'
group by property_type;'''

result4=pd.read_sql(query,conn)
result4
```

```python
# 5- Identify properties that have a higher price than the average price of properties with the same number of
#bedrooms and bathrooms.
query='''
SELECT p1.id, p1.bedrooms, p1.bathrooms, p2.price
FROM property_Details_table p1
INNER JOIN Property_price_details_table p2 ON p1.id = p2.id
WHERE p2.price > (
    SELECT AVG(price)
    FROM Property_price_details_table
    WHERE bedrooms = p1.bedrooms
    AND bathrooms = p1.bathrooms
);'''

result5=pd.read_sql(query,conn)
result5
```

```python
 # 6- Calculate the cumulative price for each property type, ordered by the creation date.
query='''SELECT
    ppd.property_type,
    pd.created_on,
    SUM(ppd.price) OVER (PARTITION BY ppd.property_type ORDER BY pd.created_on) AS cumulative_price
FROM
    property_details_table pd
INNER JOIN
    Property_price_details_table ppd
ON
    pd.id = ppd.id
ORDER BY
    ppd.property_type, pd.created_on;
'''
results6=pd.read_sql(query,conn)
results6
```

```python
# 7. SQL query to identify the 10 locations with the highest total surface area for properties listed for sale
query = '''
with subquery as(select l3, sum(surface_total) as surface_sum,operation_type
from property_details_table as pd
join Property_price_details_table as pp on pd.id=pp.id
where operation_type = 'Venta' group by l3)
select * from subquery order by surface_sum desc limit 10;

'''
result7 = pd.read_sql(query, conn)
result7
```

```python
# 8- Find the top 5 most expensive properties (based on price) in the "Palermo" neighborhood (l3) that were listed
#in August 2020:

query='''
select pd.id,price,created_on
from property_details_table as pd
join property_price_details_table as pp on pd.id=pp.id where l3='Palermo'
and created_on like '8%2020' order by price desc limit 5;
'''

result8=pd.read_sql(query,conn)
result8
```

```python
#9 Find the top 3 properties with the highest price per square meter (price divided by surface area) within each property type.

query='''
with subquery as (select pd.id,price,property_type,
rank() over(partition by property_type order by round((price/surface_total),2)desc) as rank_,
round((price/surface_total),2) as price_per_sqrmtr
from property_details_table as pd
join property_price_details_table as pp on pd.id=pp.id )
select * from subquery  where rank_ between 1 and 3;
'''

result9=pd.read_sql(query,conn)
result9
```

```python
# 10 Find the top 3 locations (l1, l2, l3) with the highest average price per square meter (price / surface_total) for properties listed for sale (operation_type = 'Venta') in the year 2020. Exclude locations with fewer than 10 properties listed for sale in 2020 from the results.

query='''
with subquery as(select l1,l2,l3,count(*)  as count_,
avg(price/surface_total) over(partition by l3) as avg_priceper_sqmtr ,end_date
from property_details_table as p1
join property_price_details_table as p2
on p1.id=p2.id
where operation_type='Venta' and end_date like'%2020' group by l1,l2,l3)
select * from subquery where count_>=10 order by avg_priceper_sqmtr desc limit 3;
'''

result10= pd.read_sql(query,conn)
result10
```
